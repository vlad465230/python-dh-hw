# Задание №1

Скачайте текст "Литературных анекдотов". Напишите функцию, которая будет читать файл, лемматизировать текст с помощью pymystem3 и записывать результат в новый файл. У функции должно бы два аргумента: путь к исходному файлу и путь к файлу с лемматизированным текстом. Вызов функции тоже должен быть прописан в решении.

```python
with open('/Users/vladislavturin/Desktop/Литературные анекдоты', 'r', encoding = 'utf-8') as a:
    text = a.read() #читаем файл и записываем текст из файла в переменную
a.close()
```

```python
from pymystem3 import Mystem

lem = Mystem()
lemmas = lem.lemmatize(text) #лемматизируем текст с помощью pymystem3

clean_lemmas = [w for w in lemmas if w.isalpha()] #очищаем текст от пунктуации и возможных чисел

lemmas1=' '.join(clean_lemmas) #объединяем лемматизированный текст в строку через пробел
print(lemmas1)

```

```python
with open('/Users/vladislavturin/Desktop/Python/python-dh-hw/Lemmas', 'w', encoding='utf-8') as f: 
    f.write(lemmas1) #записываем результат лемматизации в новый файл 
```
